```python
rows = data.shape[0]
for i in range(rows):
    price_str = data.at[i, "item_price"]
    if type(price_str) == str:
        price_str = price_str[1:]
        price = float(price_str)
    
        data.at[i, "item_price"] = price
data["item_price"] = data["item_price"].astype(float)
```
